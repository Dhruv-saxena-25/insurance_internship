```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (15,15))

s= sns.countplot(data= df, x= 'region', ax= axes[0][0])
for p in s.patches:
   s.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))


s= sns.countplot(data= df, x= 'sex', ax= axes[0][1])
for p in s.patches:
   s.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

s= sns.countplot(data= df, x= 'children', ax= axes[1][0])
for p in s.patches:
   s.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

s= sns.countplot(data= df, x= 'smoker', ax= axes[1][1])
for p in s.patches:
   s.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

plt.show()
```

- Maximum number of people are from `southwest` region followed by `northwest and southwest` region has same number of peoples.
- Minimum number of people are from `northeast` region.

```python
ax = df.groupby(['region'])['expenses'].max().plot(kind= 'bar')
for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x(), p.get_height()+0.01))

plt.show()
```

- Maximum amount of expenses in `northeast` region is `58571.07`.
- Maximum amount of expenses in `northwest` region is `60021.40`.
- Maximum amount of expenses in `southeast` region is `60021.40`.
- Maximum amount of expenses in `southwest` region is `52590.83`.


```python
ax= df.groupby(['region'])['expenses'].min().plot(kind= 'bar')
for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.05, p.get_height()+0.01))

plt.show()
```

- Minimum amount of expenses in `northeast` region is `1694.80`.
- Minimum amount of expenses in `northwest` region is `1621.34`.
- Minimum amount of expenses in `southeast` region is `1121.87`.
- Minimum amount of expenses in `southwest` region is `1241.57`.

```python
## Here, we are plotting violin plot realted with x = sex & y = Expenses with HUe function on "Smoker" Columns.

palette=['#EB5050','#2DFFAB'] 
sns.catplot(x='sex', y='expenses', kind='violin', hue='smoker', palette=palette, data=df)
```

```python
## here we are plotting scatterplot with x = bmi & y = expenses & relationship with hue functions on [ sex, smoker, regions].

palette=['#EB5050','#3EA2FF']
fig, axes = plt.subplots(ncols = 3, figsize = (15,6), squeeze=True)
sns.scatterplot(x='bmi', y='expenses', ax=axes[0], data=df,hue='sex', palette=palette)
sns.scatterplot(x='bmi', y='expenses', ax=axes[1], data=df,hue='smoker', palette=palette)
sns.scatterplot(x='bmi', y='expenses', ax=axes[2], data=df,hue='region')
```

```python
# Here, we are plotting Distribution- plot on Expenses columns and going checking distribution related with [ mean, std,]

from scipy import stats
from scipy.stats import norm
fig =plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.distplot(df['expenses'], fit=norm)
(mu,sigma)= norm.fit(df['expenses'])
plt.legend(['For Normal dist. mean: {:.2f} | std: {:.2f}'.format(mu,sigma)])
plt.ylabel('Frequency')
plt.title('Distribution of Charges')
```
